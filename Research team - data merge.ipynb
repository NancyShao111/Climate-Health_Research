```python
import pandas as pd
```

```python
emission = pd.read_csv('~/Desktop/emission data.csv')
laws = pd.read_csv('~/Desktop/laws and incentives - final.csv')
state_name = pd.read_csv('~/Desktop/state_name.csv')
```

```python
emission.head()
```

```python
state_name.head()
```

```python
laws.head()
```

```python
laws = laws.rename(columns = {'YEAR':'Year'})
```

```python
emission_new = pd.merge(
                     emission,
                     state_name,
                     how  = 'left', on = 'State')
emission_new = emission_new.drop('State', axis=1)
emission_new = emission_new.rename(columns = {'Abbreviation':'STATE'})
cols = emission_new.columns.tolist()
cols = cols[-1:] + cols[:-1]
emission_new = emission_new[cols]
```

```python
emission_new.head()
```

```python
df_final = pd.merge(
                     emission_new,
                     laws,
                     how  = 'left', on = ['STATE','Year'])
```

```python
df_final.to_csv('~/Desktop/final data.csv')
```
